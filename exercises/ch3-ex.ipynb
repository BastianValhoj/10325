## a)

Time-dependent Schdinger equation (TDSE):
$$
    \mathbf{H}\phi(t) = i\frac{\partial}{\partial t}\phi(t),\quad \phi(t=0) = f
$$

We want to show that 
$$
    \phi(t) = \mathbf{G}(t)f, \quad \mathbf{G}(t=0) = \mathbf{1}
$$

Is a solution to TDSE. Now, since $f$ is time-independent we can write the equation as

$$\begin{align*}
    \mathbf{H}\phi(t)   &= i\frac{\partial}{\partial t}\phi(t)
    \\
    \mathbf{H}\Big[\mathbf{G}(t)f\Big] &= i\left[\frac{\partial}{\partial t}\mathbf{G}(t)f\right]
    \\
    \mathbf{H}\Big[\mathbf{G}(t)\Big]f &= i\left[\frac{\partial}{\partial t}\mathbf{G}(t)\right] f
    \\
    \mathbf{H}\mathbf{G}(t) &= i\frac{\partial}{\partial t}\mathbf{G}(t)
\end{align*}$$

Since $f$ is simply a scaling "factor" (a vector really), and is independent of time-variables, usign $\phi(t) = \mathbf{G}(t)f$ is equivalent to scaling both sides of TDSE by a (vector) constant.  
The initial conditions are also satisfied:

$$\begin{align*}
    \mathbf{G}(t=0) &= \mathbf{1}
    \\
    \phi(t=0) &= f
    \\
    \to \phi(t=0) &= \mathbf{G}(t=0) f = \mathbf{1}f 
    \\ &= 
            \begin{pmatrix} 
                f_1 & 0     &  \cdots & &\\
                0   & f_2   &         & &\\
                \vdots & &  \ddots &  &\\ 
                \\
            \end{pmatrix}
\end{align*}$$


## b)

The initial condition of the greens matrix $\mathbf{G}(t=0)$ is a identity matrix. Then $f$ corresponds to the initial value of each system.

Solving the TDSE, Greens function must then correspond to a superposition of the solutions to TDSE using the intial values of $f$.

Wee need to prove that 
$$
    \mathbf{G} = e^{-i\mathbf{H}t} \equiv \sum_{n}^\infty \frac{\left(-i\mathbf{H}t\right)^n}{n!}
$$

is a solution to 
$$
    \mathbf{HG}(t) = i\frac{\partial }{\partial t}\mathbf{G}(t).
$$

We simply insert the expression in the differential equation:

$$\begin{align*}
    \mathbf{HG}(t) &= i\frac{\partial }{\partial t}\mathbf{G}(t) 
    \\ &= i\frac{\partial }{\partial t}e^{-i\mathbf{H}t}
    \\ &= -i^2\mathbf{H}e^{-i\mathbf{H}t}
    \\ &= \mathbf{HG}(t) \quad \square
\end{align*}$$
