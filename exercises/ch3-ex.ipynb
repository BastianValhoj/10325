## a)

Time-dependent Schödinger equation (TDSE):
$$
    \mathbf{H}\phi(t) = i\frac{\partial}{\partial t}\phi(t),\quad \phi(t=0) = f
$$

We want to show that 
$$
    \phi(t) = \mathbf{G}(t)f, \quad \mathbf{G}(t=0) = \mathbf{1}
$$

Is a solution to TDSE. Now, since $f$ is time-independent we can write the equation as

$$\begin{align*}
    \mathbf{H}\phi(t)   &= i\frac{\partial}{\partial t}\phi(t)
    \\
    \mathbf{H}\Big[\mathbf{G}(t)f\Big] &= i\left[\frac{\partial}{\partial t}\mathbf{G}(t)f\right]
    \\
    \mathbf{H}\Big[\mathbf{G}(t)\Big]f &= i\left[\frac{\partial}{\partial t}\mathbf{G}(t)\right] f
    \\
    \mathbf{H}\mathbf{G}(t) &= i\frac{\partial}{\partial t}\mathbf{G}(t)
\end{align*}$$

Since $f$ is simply a scaling "factor" (a vector really), and is independent of time-variables, usign $\phi(t) = \mathbf{G}(t)f$ is equivalent to scaling both sides of TDSE by a (vector) constant.  
The initial conditions are also satisfied:

$$\begin{align*}
    \mathbf{G}(t=0) &= \mathbf{1}
    \\
    \phi(t=0) &= f
    \\
    \to \phi(t=0) &= \mathbf{G}(t=0) f = \mathbf{1}f 
    \\ &= 
            \begin{pmatrix} 
                f_1 & 0     &  \cdots & &\\
                0   & f_2   &         & &\\
                \vdots & &  \ddots &  &\\ 
                \\
            \end{pmatrix}
\end{align*}$$


## b)

The initial condition of the greens matrix $\mathbf{G}(t=0)$ is a identity matrix. Then $f$ corresponds to the initial value of each system.

Solving the TDSE, Greens function must then correspond to a superposition of the solutions to TDSE using the intial values of $f$.

Wee need to prove that 
$$
    \mathbf{G} = e^{-i\mathbf{H}t} \equiv \sum_{n}^\infty \frac{\left(-i\mathbf{H}t\right)^n}{n!}
$$

is a solution to 
$$
    \mathbf{HG}(t) = i\frac{\partial }{\partial t}\mathbf{G}(t).
$$

We simply insert the expression in the differential equation:

$$\begin{align*}
    \mathbf{HG}(t) &= i\frac{\partial }{\partial t}\mathbf{G}(t) 
    \\ &= i\frac{\partial }{\partial t}e^{-i\mathbf{H}t}
    \\ &= -i^2\mathbf{H}e^{-i\mathbf{H}t}
    \\ &= \mathbf{HG}(t) \quad \square
\end{align*}$$

To recap: eq. (30), (33), and (37) is:

$$
    i\frac{\partial}{\partial t}\mathbf{G}(t)  - \mathbf{HG}(t) = \delta(t)\mathbf{1} \tag{30}
$$

$$
    \left[\left(E + i\eta\right)\mathbf{I} - \mathbf{H}\right]\mathbf{G}(E) = \mathbf{1} \tag{33}
$$

$$
    \mathbf{G}(E) = \lim_{\eta\to 0_+}\int_{-\infty}^\infty dt\     e^{i(E + i\eta)t}\mathbf{G}(t) \tag{37}
$$

$$\begin{align*}
    i\frac{\partial}{\partial t}\mathbf{G}(t) - \mathbf{H}\mathbf{G}(t) &= \delta(t)\mathbf{1}
    \\ 
    \int_{-\infty}^\infty dt\ e^{i(E + i\eta)t} i\frac{\partial}{\partial t}\mathbf{G}(t) - \int_{-\infty}^\infty dt\ e^{i(E + i\eta)t}\mathbf{H}\mathbf{G}(t) &= \int_{-\infty}^\infty dt\ e^{i(E+i\eta)}\delta(t)\mathbf{1}
    \\
\end{align*}$$

Now we can use integration by parts for the LHS above
$$\begin{align*}
    \partial_t uv &= u\partial_t v + v\partial_t u 
    \\
    \int_{a}^b dt\ \partial_t uv &= \int_a^b dt\ u\partial_t v + \int_a^b dt\ v\partial_t u
    \\
    [uv]^b_a &= \int_a^b dt\ u\partial_t v + \int_a^b dt\ v\partial_t u
    \\ \Rightarrow \int_a^b dt\ u\partial_t v &= [uv]^b_a - \int_a^b dt\ v\partial_t u
    % \mathbf{1} &= [(E - i\eta)\mathbf{I} - \mathbf{H}]\mathbf{G}(E) = [(E - i\eta)\mathbf{I} - \mathbf{H}]\left(\lim_{\eta\to0_+}\int_{-\infty}^\infty{dt}\ e^{i(E+i\eta)t}\mathbf{G}(t)\right) 
    % \\ &= (E - i\eta)\mathbf{I}\lim_{\eta\to0_+}\int_{-\infty}^\infty{dt}\ e^{i(E+i\eta)t}\mathbf{G}(t) - \mathbf{H}\lim_{\eta\to0_+}\int_{-\infty}^\infty{dt}\ e^{i(E+i\eta)t}\mathbf{G}(t)
    % \\
    % \lim_{\eta_2\to0_+}\frac{1}{2\pi}\int_{-\infty}^\infty dt\ e^{-i(E-i\eta_2)t}\mathbf{1} &= \frac{1}{2\pi}\lim_{\eta_2 \to 0_+}\int_{-\infty}^\infty e^{-i(E-i\eta_2)t}\left(\lim_{\eta_1\to0_+}\int_{-\infty}^\infty dt\ e^{i(E+i\eta_1)}\mathbf{G}(t)\right)
    % \\
    % \lim_{\eta_2\to0_+}\delta(E-i\eta_2)\mathbf{1} &= \frac{1}{2\pi}\lim_{\eta_2 \to 0_+}\int_{-\infty}^\infty e^{-i(E-i\eta_2)t}\left(\lim_{\eta_1\to0_+}\int_{-\infty}^\infty dt\ e^{i(E+i\eta_1)}\mathbf{G}(t)\right)
\end{align*}$$

And so we have

$$\begin{align*}
    \int_{-\infty}^\infty dt\ e^{i(E + i\eta)t} i\frac{\partial}{\partial t}\mathbf{G}(t) - \int_{-\infty}^\infty dt\ e^{i(E + i\eta)t}\mathbf{H}\mathbf{G}(t) &= \int_{-\infty}^\infty dt\ e^{i(E+i\eta)}\delta(t)\mathbf{1}
    \\
    i\left(
        \left[e^{i(E+i\eta)t} \mathbf{G}(t)\right]^\infty_{-\infty} - \int_{-\infty}^\infty dt\ \mathbf{G}(t) \frac{\partial }{\partial t}e^{i(E+i\eta)t}
    \right)
     - \mathbf{H}\int_{-\infty}^\infty dt\ e^{i(E + i\eta)t}\mathbf{G}(t) &= \mathbf{1}
\end{align*}$$
Since $\lim_{t\to\infty} e^{(iE - \eta)t} = 0$, and  $\mathbf{G}(t) = \left(-i\theta(t) e^{-i\mathbf{H}t}\right)$, the limit of $\lim_{t\to-\infty}  \mathrm{G}(t) = 0$, the LHS is zero.

$$\begin{align*} 
     i\left(
        0 - i(E+i\eta)\int_{-\infty}^\infty dt\ e^{i(E+i\eta)t}\mathbf{G}(t)
    \right)
     - \mathbf{H}\int_{-\infty}^\infty dt\ e^{i(E + i\eta)t}\mathbf{G}(t) &= \mathbf{1}
     \\
     (E+i\eta)\int_{-\infty}^\infty dt\ \mathbf{G}(t) - \mathbf{H}\int_{-\infty}^\infty dt\ e^{i(E+i\eta)t}\mathbf{G}(t) = \mathbf{1}
\end{align*}$$

Now, since $\eta$ is very small, we change the fourier transform to check the limit of $\eta\to0_+$

$$\begin{align*}
    (E+i\eta)\lim_{t\to 0_+}\int_{-\infty}^{\infty} dt\ e^{i(E+i\eta)t}\mathrm{G}(t) - \mathbf{H}\lim_{\eta \to 0_+}\int_{-\infty}^\infty e^{i(E+i\eta)t}\mathrm{G}(t) &= \mathrm{1}
    \\
    (E+i\eta)\mathbf{G}(E) - \mathbf{H}\mathrm{G}(E) &= \mathrm{1}
    \\
    \left[ (E+i\eta)\mathbf{I}- \mathrm{H}\right]\mathbf{G}(E) &= \mathrm{1} \quad \square
\end{align*}$$
